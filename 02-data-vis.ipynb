```python
data = df.loc[(df["SKU"] == 32485) & (df["geoCluster"] == 2043)]

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=data['date'], 
    y=data["sales"],
    mode="lines",
    connectgaps=True,
    line=dict(color="rgb(67,67,67)")
))


fig.show()
```

```python
selected_sku = 32485   # Banana
geos = [2043, 1935, 2126]

fig = make_subplots(
    rows=3, cols=2, 
    subplot_titles=(
        f"Banana sales in geo cluster {geos[0]}", 
        f"Banana prices in geo cluster {geos[0]}", 
        f"Banana sales in geo cluster {geos[1]}", 
        f"Banana prices in geo cluster {geos[1]}",
        f"Banana sales in geo cluster {geos[2]}", 
        f"Banana prices in geo cluster {geos[2]}",
    )
)

# row 1
data = df.loc[(df["SKU"] == selected_sku) & (df["geoCluster"] == geos[0])]

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["sales"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(67,67,67)")
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["price"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(115,115,115)")
    ),
    row=1, col=2
)

# row 2
data = df.loc[(df["SKU"] == selected_sku) & (df["geoCluster"] == geos[1])]

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["sales"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(67,67,67)"),
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["price"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(115,115,115)"),
        name="Name of Trace 1"
    ),
    row=2, col=2
)

# row 3
data = df.loc[(df["SKU"] == selected_sku) & (df["geoCluster"] == geos[2])]

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["sales"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(67,67,67)"),
    ),
    row=3, col=1
)

fig.add_trace(
    go.Scatter(
        x=data['date'], 
        y=data["price"],
        mode="lines",
        connectgaps=True,
        line=dict(color="rgb(115,115,115)"),
    ),
    row=3, col=2
)


fig.update_layout(height=600, width=1000, title_text="Subplots")
for i in fig['layout']['annotations']:
    i['font'] = dict(size=10,color='rgb(67,67,67)')


fig.show()
```
